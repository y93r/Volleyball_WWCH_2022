```python
#Bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (15, 5)  # Tamanho do gráfico
plt.style.use('seaborn-darkgrid') # Estilo do gráfico
```

```python
brasil = pd.read_csv(r'Brasil.csv',sep= ',',encoding='utf-8',skipinitialspace = True)
eua = pd.read_csv(r'Eua.csv',sep= ',',encoding='utf-8',skipinitialspace = True)
italia = pd.read_csv(r'Italia.csv',sep= ',',encoding='utf-8',skipinitialspace = True)
servia = pd.read_csv(r'Servia.csv',sep= ',',encoding='utf-8',skipinitialspace = True)
china = pd.read_csv(r'China.csv',sep= ',',encoding='utf-8',skipinitialspace = True)

display(brasil)
display(italia)
display(eua)
display(servia)
display(china)
```

```python
#concatenar as tabelas
df = pd.concat([eua, italia,brasil,servia,china])
df = df.reset_index(drop=True)
df_group = df.groupby(list(df.columns))
print(df)
```

```python
#comparativo média de idade e altura | Time com maior média de idade
df.groupby('PAÍS')[['IDADE','ALTURA']].mean().round(2).sort_values('IDADE', ascending=False)
```

```python
#gráficos
df.groupby('PAÍS')['IDADE'].mean()
media_idade = sns.barplot(x='PAÍS', y='IDADE', data=df, ci=None)
media_idade.set_ylim(20,30)
media_idade.set_title('Média Idade por seleção', fontsize = 16)
plt.show()
```

```python
idade = df['IDADE']
sns.histplot(data=df, x=idade, kde=True).set_title(f"Distribuição da variável Idade", fontsize = 16)
plt.show()
```

```python
#comparativo média de idade e altura | Time com maior média de altura
df.groupby('PAÍS')[['ALTURA', 'IDADE']].mean().round(2).sort_values('ALTURA', ascending=False)
```

```python
#gráficos
df.groupby('PAÍS')['ALTURA'].mean()
media_idade = sns.barplot(x='PAÍS', y='ALTURA', data=df, ci=None)
media_idade.set_ylim(1.60, 1.92)
media_idade.set_title('Média altura por seleção', fontsize = 16)
plt.show()
```

```python
altura = df['ALTURA']
sns.histplot(data=df, x=altura , kde=True).set_title(f"Distribuição da variável Altura", fontsize = 16)
plt.show()
```

```python
#ordenar pelas 10 mais velhas
df.sort_values("IDADE", axis = 0, ascending = False,
                    inplace = True, na_position ='last') 
print(df.head(10))
```

```python
#ordenar pelas 10 mais novas
df.sort_values("IDADE", axis = 0, ascending = True,
                    inplace = True, na_position ='last') 
print(df.head(10))
```

```python
#ordenar pelas 10 mais altas
df.sort_values("ALTURA", axis = 0, ascending = False,
                    inplace = True, na_position ='last') 
print(df.head(10))
```

```python
#ordenar pelas 10 mais baixas
df.sort_values("ALTURA", axis = 0, ascending = True,
                    inplace = True, na_position ='last') 
print(df.head(10))
```

```python
# Unicos 
#Posição tem quer ser igual a 5(Ponteira,Meio de rede,Oposto,Líbero)
df.nunique()
```

```python
#descobrir qual valor está diferente
df['POSIÇÃO'].value_counts()
```

```python
#Substituir os valores da coluna posição que estão escritos diferente
df.loc[ df['POSIÇÃO'] == 'Meio De Rede' ].head()
```

```python
df['POSIÇÃO'].replace(["Meio De Rede"], "Meio de rede", inplace = True)
```

```python
# Analise idade por posição
analise_idade = df.groupby( by=['POSIÇÃO'] ).agg(
    Quantidade = ('IDADE', 'count'),
    Media = ('IDADE', 'mean'),
    Max = ('IDADE', 'max'),
    Min = ('IDADE', 'min'),
    Mediana = ('IDADE', 'median'),
)


analise_idade.sort_values('Media', ascending=False ).round(2)
```

```python
# Analise idade por país
analise_idade_pais = df.groupby( by=['PAÍS'] ).agg(
    Quantidade = ('IDADE', 'count'),
    Media = ('IDADE', 'mean'),
    Max = ('IDADE', 'max'),
    Min = ('IDADE', 'min'),
    Mediana = ('IDADE', 'median'),
)


analise_idade_pais.sort_values('Media', ascending=False ).round(2)
```

```python
# Analise altura por posição
analise_altura = df.groupby( by=['POSIÇÃO'] ).agg(
    Quantidade = ('ALTURA', 'count'),
    Media = ('ALTURA', 'mean'),
    Max = ('ALTURA', 'max'),
    Min = ('ALTURA', 'min'),
    Mediana = ('ALTURA', 'median'),
)


analise_altura.sort_values('Media', ascending=False ).round(2)
```

```python
# Analise altura por altura
analise_altura_pais = df.groupby( by=['PAÍS'] ).agg(
    Quantidade = ('ALTURA', 'count'),
    Media = ('ALTURA', 'mean'),
    Max = ('ALTURA', 'max'),
    Min = ('ALTURA', 'min'),
    Mediana = ('ALTURA', 'median'),
)


analise_altura_pais.sort_values('Media', ascending=False).round(2)
```
